```python
import os

from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document

import faiss

from utils import get_llm_instance, load_pdf
```

```python
PERSISTENT_VECTOR_STORE_NAME = "persistent_vector_store"
PAPERS_DIR = "papers/"
```

```python
# Set up vector store
embeddings = get_llm_instance("main_embedding_model", embedding=True)
embedding_dim = len(embeddings.embed_query("Not with a bang but a whimper."))
index = faiss.IndexFlatL2(embedding_dim)

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
papers = []

for area in os.listdir(PAPERS_DIR):
    for paper in os.listdir(f"{PAPERS_DIR}{area}"):
        title, chunks = load_pdf(f"{PAPERS_DIR}{area}/{paper}")
        papers.append((title, area, chunks))
```

```python
texts, metadatas = [], []

for title, area, chunks in papers:
    for i, chunk in enumerate(chunks):
        texts.append(chunk)
        metadatas.append({
                "title": title+f" [CHUNK{i}]",
                "area": area
                })
    
vector_store.add_texts(texts=texts, metadatas=metadatas)
```

```python
vector_store.save_local(PERSISTENT_VECTOR_STORE_NAME)
```

```python
for metadata in metadatas:
    print(metadata)
```
