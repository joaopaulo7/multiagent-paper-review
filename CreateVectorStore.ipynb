```python
import faiss

from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document

from utils import get_llm_instance
```

```python
# Set up vector store
embeddings = get_llm_instance("main_embedding_model", embedding=True)
embedding_dim = len(embeddings.embed_query("Not with a bang but a whimper."))
index = faiss.IndexFlatL2(embedding_dim)

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
doc = Document(page_content="Bunda", metadata={"title": "teste1", "area": "none1"})

vector_store.add_documents([doc])
```

```python
vector_store.save_local("vector_store")
```
